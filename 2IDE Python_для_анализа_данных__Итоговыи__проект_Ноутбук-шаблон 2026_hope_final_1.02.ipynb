3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)

* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.

```python
def experience(text):
    if pd.isna(text) or text == "Не указано":
        return np.nan

    text = str(text).lower()

    if "опыт работы" in text:
        text = text.split("опыт работы")[1].strip()

    words = text.split()[:4]

    years = 0
    months = 0

    for i, w in enumerate(words):
        if w.isdigit():
            num = int(w)
            if i+1 < len(words):
                next_word = words[i+1]
                if next_word in ["год", "года", "лет"]:
                    years = num
                elif next_word in ["месяц", "месяца", "месяцев"]:
                    months = num

    return years * 12 + months
```

```python
df["Опыт работы (месяц)"] = df["Опыт работы"].apply(experience)
df[["Опыт работы", "Опыт работы (месяц)"]].describe()
```

```python
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань',
                  'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа',
                  'Красноярск', 'Пермь', 'Воронеж','Волгоград']

def extract_city(text):
    if pd.isna(text):
        return "другие"

    city = text.split(",")[0].strip()
    if city == "Москва":
        return "Москва"
    elif city == "Санкт-Петербург":
        return "Санкт-Петербург"
    elif city in million_cities:
        return "город-миллионник"
    else:
        return "другие"
```

```python
def extract_relocation(text):
    if pd.isna(text):
        return False

    text = text.lower()
    if "не готов к переезду" in text or "не готова к переезду" in text:
        return False
    elif "готов к переезду" in text or "готова к переезду" in text or "хочу переехать" in text:
        return True
    else:
        return False
```

```python
def extract_travel(text):
    if pd.isna(text):
        return False

    text = text.lower()
    if "не готов к командировкам" in text or "не готова к командировкам" in text or "не готова к редким командировкам" in text:
        return False
    elif "готов к командировкам" in text or "готова к командировкам" in text or "готов к редким командировкам" in text or "готова к редким командировкам" in text:
        return True
    else:
        return False
```

```python
df["Город"] = df["Город, переезд, командировки"].apply(extract_city)
df["Готовность к переезду"] = df["Город, переезд, командировки"].apply(extract_relocation)
df["Готовность к командировкам"] = df["Город, переезд, командировки"].apply(extract_travel)

# df.drop(columns=["Город, переезд, командировки"], inplace=True)
df.head(2)
```

На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>

```python
#ваш код здесь
employment_categories = [
    "полная занятость",
    "частичная занятость",
    "проектная работа",
    "стажировка",
    "волонтерство"
]
df_employment = df["Занятость"].str.get_dummies(sep=", ").astype(bool)
df = pd.concat([df, df_employment], axis=1)

schedule_categories = [
    "полный день",
    "сменный график",
    "гибкий график",
    "удаленная работа",
    "вахтовый метод"
]

df_schedule = df["График"].str.get_dummies(sep=", ").astype(bool)
df = pd.concat([df, df_schedule], axis=1)

df.drop(columns=["Занятость", "График"], inplace=True)

```

```python
rates = pd.read_csv('ExchangeRates.csv')
rates = rates[['currency', 'date', 'proportion', 'close']]
rates['date'] = pd.to_datetime(rates['date']).dt.date
df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме'], dayfirst=True)
df['resume_date'] = df['Обновление резюме'].dt.date
df['salary_amount'] = (
    df['ЗП']
    .str.replace('\xa0', '', regex=False)
    .str.extract(r'(\d+)')
    .astype(float)
)

df['salary_currency_raw'] = df['ЗП'].str.extract(r'([^\d\s]+)')
currency_map = {
    'руб.': 'RUB',
    'руб': 'RUB',
    'RUR': 'RUB',
    'USD': 'USD',
    'EUR': 'EUR',
    'грн': 'UAH',
    'белруб': 'BYN',
    'KGS': 'KGS',
    'сум': 'UZS',
    'AZN': 'AZN',
    'KZT': 'KZT'
}

df['currency'] = df['salary_currency_raw'].map(currency_map)
df = df.merge(
    rates,
    how='left',
    left_on=['currency', 'resume_date'],
    right_on=['currency', 'date']
)
df.loc[df['currency'] == 'RUB', ['close', 'proportion']] = 1
df['ЗП (руб)'] = df['salary_amount'] * df['close'] / df['proportion']
```
